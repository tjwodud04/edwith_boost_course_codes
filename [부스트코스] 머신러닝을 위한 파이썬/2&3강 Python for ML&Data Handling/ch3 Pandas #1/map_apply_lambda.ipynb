### apply & applymap

```python
df = pd.read_csv("wages.csv")
df.head()
```

```python
df_info = df[["earn", "height","age"]]
df_info.head()
```

```python
f = lambda x : x.max() - x.min()
df_info.apply(f)
```

```python
df_info.apply(sum)
```

```python
df_info.sum()
```

```python
def f(x):
    return Series([x.min(), x.max(), x.mean()], index=["min", "max", "mean"])

df_info.apply(f)
```

```python
f = lambda x : -x

df_info.applymap(f).head(5)
```

```python
f = lambda x : -x

df_info["earn"].apply(f).head(5)
```
